### 8. Tenure Calculation
Calculate how many days each employee has worked since their `HireDate`. You will need to convert `HireDate` to datetime objects first.

```python
# Your code here
df['HireDate'] = pd.to_datetime(df['HireDate'])

```

```python
today = pd.Timestamp.today()

df['Tenure_Years'] =  (today - df['HireDate']).dt.days / 365.25

print(df['Tenure_Years'])
```
