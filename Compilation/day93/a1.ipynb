```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('sales_acquisition_200.csv')
```

```python
df.head(10)
```

```python
total_revenue = df['Expected_Revenue'].sum()
```

```python
print('Total Expected Revenue: $',total_revenue)
```

```python
#search leads where comming

df['Source'].value_counts()
```

```python
#CATCH WINNING DEAL

won_deals = df[df['Status']=='Closed Won']
won_revenue = won_deals['Expected_Revenue'].sum()

```

```python
webinar_leads = df[df['Source'] == 'Webinar']

avg_cost = webinar_leads['Acquisition_Cost'].sum()
print('Average Webinar Cost is: $',avg_cost)
```

```python
cost_by_source = won_deals.groupby('Source')['Acquisition_Cost'].sum()

print(cost_by_source)
```

```python
df['Priority'] = np.where(df['Expected_Revenue'] > 10000,'High Value','Standard')

deal_counts = df['Priority'].value_counts()
print(deal_counts)
```
