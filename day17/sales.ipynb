```python
import pandas as pd
```

```python
data = {
    'Date': ['2025-01-01','2025-01-02','2025-01-03','2025-01-04',
             '2025-01-05','2025-01-06','2025-01-07','2025-01-08'],
    'Region': ['North','North','South','East','West','South','East','North'],
    'Product': ['Whey','Creatine','Mass Gainer','Whey','Preworkout','Whey','Creatine','Mass Gainer'],
    'Salesperson': ['Alex','Ben','Cara','Dana','Evan','Cara','Dana','Ben'],
    'Units Sold': [10,8,5,12,7,6,9,4],
    'Unit Price': [500,450,900,500,750,500,450,900]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Total Sales"] = df['Units Sold'] * df['Unit Price']
print(df)
```

```python
pivot = pd.pivot_table(df,
                       values='Total Sales',
                       index='Region',
                       columns='Product',
                       aggfunc='sum',
                       fill_value=0)

print(pivot)
```

```python
pivot['Total'] = pivot.sum(axis=1)
pivot = pivot.sort_values('Total',ascending=False)
print(pivot)
```
