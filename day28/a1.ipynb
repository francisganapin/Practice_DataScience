```python
sales_amount = 150000

steven = {'commision':0}
miguel = {'commision':0}


if sales_amount > 15000:
    print('Sales targeted exceeded bonus awarded: 1000')
    steven['commision'] += 1000
    miguel['commision'] += 1000
    print(steven,miguel)
elif sales_amount > 10000 and sales_amount <= 15000:
    print('Sales targeted between 10000 and 15000 bonus awarded: 500')
    steven['commision'] += 500
    miguel['commision'] += 500
    print(steven,miguel)
elif sales_amount  < 9999 < sales_amount <= 5000:
    print('Sales targeted between 5000 and 9999 bonus awarded: 200')
    steven['commision'] += 200
    miguel['commision'] += 200
    print(steven,miguel)
else:
    print('sales target not met. keep pushing!')
    steven['commision'] += 50
    miguel['commision'] += 50
    print(steven,miguel)
```

```python
purchase_amount = 800

is_premium_member = True

if purchase_amount > 200 and is_premium_member:
    discount_percent = 0.2
    print(f"Premium member discount applied: {discount_percent*100}%")
elif purchase_amount > 200:
    discount_percent = 0.1
    print(f'standard discount applied: {discount_percent*100}%')
elif purchase_amount >= 100:
    discount_percent = 0.05
    print(f'standard discount applied: {discount_percent*100}%')
else:
    discount_percent = 0
    print('no discount applied')

discount_amount = purchase_amount * (discount_percent / 100)
final_price = purchase_amount - discount_amount


print(f'Original Price: ${purchase_amount}')
print(f"Discount Amount: {discount_percent}% (${discount_amount})")
print(f"Final Price: ${final_price}")
```

```python
input = [3,1,4,4,2]

res = input[0]
for n in input:
    if n > res:
        res = n

print(f"The largest number in the list is: {res}")
```

```python
def count_vowels(s):
    vowels = 'aeiou'
    count = 0
    s = s.lower()
    for char in s:
        if char in vowels:
            count += 1
    return count
    
    
print(count_vowels('Hello World'))
print(count_vowels('Data Programming'))
```
