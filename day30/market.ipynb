```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
df = pd.read_csv('market.csv')
```

```python
total_sales = df['Total'].sum()


print(f'The total Sales was {total_sales:.2f}')
```

```python
best_branch = df.groupby('Branch')['Total'].sum().idxmax()
print(f'The best branch was {best_branch}')
```

```python
payment_method_unique = df['Payment'].unique()
print(payment_method_unique)
```

```python
product_rating = df.groupby('ProductLine')['Rating'].mean()
print(product_rating)
```

```python
gender_analysis = df.groupby('City')['Gender'].count()
print(gender_analysis)
```

```python





total_tax_per_branch = df.groupby('Branch')['Tax'].sum()
print(total_tax_per_branch)
```

```python
member_spending_normal = df[df['CustomerType'] == 'Normal']['Total'].sum()
member_spending_member = df[df['CustomerType'] == 'Member']['Total'].sum()
print(f'Normal member spend {member_spending_normal:.2f} while Member spend {member_spending_member:.2f}')
```

```python
hourly_spending = df.groupby('Time')['Total'].sum()
print(hourly_spending)
```

```python
correlation_analysis = df.corr()['UnitPrice']['Rating']
print(correlation_analysis)
```

```python
correlation_unit_price = df['UnitPrice'].corr(df['Quantity'])
print(correlation_unit_price)
```

```python
best_selling_in_yagon = (
    df[df['City'] == 'Yangon']
    .groupby('ProductLine')['Quantity']
    .sum()
    .idxmax()
)

print(best_selling_in_yagon)
```
