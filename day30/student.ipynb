```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_performance.csv')
```

```python
average_student_age = df['Age'].mean()
print(f'Average student age: {average_student_age}')
```

```python
attendance_check = df.groupby('Name')['Attendance'].idxmax()
print(attendance_check)
```

```python
attendance_sum = df.groupby('Name')['Attendance'].sum()


top_name = attendance_sum.idxmax()
top_value = attendance_sum.max()

print(top_name,top_value)
```

```python
attendance_sum = df.groupby('Name')['Attendance'].sum()

if not attendance_sum.empty:
    print(attendance_sum.idxmax(),attendance_sum.max())
else:
    print('No Attendance data')
```

```python
study_correlation = df['StudyHours'].corr(df['MathScore'])
```

```python
df['AverageScore'] = df[['MathScore','ScienceScore','EnglishScore']].mean(axis=1)
```

```python
performance_cluster = df['AverageScore'].apply(lambda x: 'High' if x > 90 else 'Medium' if x > 75 else 'Low')

print(performance_cluster)
```

```python
age_vs_performance  = df['Age'].corr(df['AverageScore'])
print(age_vs_performance)
```
