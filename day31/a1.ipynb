```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Market.csv')
df.head()
```

```python
total_revenue_per_region = df.groupby('Region')['Amount'].sum()

print(total_revenue_per_region)
```

```python
revenue_electronic = df[df['Category'] == 'Electronics']['Amount'].sum()
revenue_furniture = df[df['Category'] == 'Furniture']['Amount'].sum()

print('Revenue of electronic ', revenue_electronic)
print('Revenue of furniture  ', revenue_furniture)
print('Revenue diffrence was ', revenue_electronic - revenue_furniture)
```

```python
top_product_from_north_region = df[df['Region'] == 'North'].groupby('Product').sum().sort_values(by='Amount',ascending=False).head(1)

print(top_product_from_north_region)
```

```python
peak_days  = df['Amount'].max().groupby(df['Date']).idxmax()
print(peak_days)
```

```python
peak_days = df.groupby('Date')['Amount'].sum().idxmax()

print(peak_days)
```

```python
daily_totals = df.groupby('Date')['Amount'].sum().idxmax()
```

```python
daily_total = df.groupby('Date')['Amount'].sum()
peak_days = daily_total[daily_total == daily_total.max()]

print(peak_days)
```

```python
df['Transaction'] = df['Amount'].apply(lambda x: 'High' if x > 500 else 'Low')
```

```python
print(df)
```

```python
order_most = df['Product'].value_counts()
print(order_most)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month_name()

print(df)
```

```python
underperforming_region = df.groupby('Region').sum().sort_value(by='Amount',ascending=False)
print(underperforming_region)
```
