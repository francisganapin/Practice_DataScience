Problem 15: Sold vs Active
How many properties are Sold vs Active? Show as percentages.

```python
active_properties = df[df['status'] == 'Active']
sold_properties = df[df['status'] == 'Sold']




total = len(active_properties) + len(sold_properties)

active_pct = (len(active_properties)/total) * 100
sold_pct = (len(sold_properties)/total) * 100


print(f"Active: {len(active_properties)} ({active_pct:.2f}%)")
print(f"Sold: {len(sold_properties)} ({sold_pct:.2f}%)")



```

Problem 17: Price Difference
For sold properties, calculate the difference between list_price and sale_price. Create a new column price_difference.

```python
sold_properties = df[df['status'] == 'Sold'].copy()

sold_properties['price_difference'] = (
    sold_properties['sale_price'] - sold_properties['list_price']
)

sold_properties[[
    'property_id',
    'address',
    'city',
    'list_price',
    'sale_price',
    'price_difference'
]].head(5)

```

Problem 20: Agent Commission Earned
For sold properties, calculate the actual commission earned (sale_price  agent_commission) and show total commission per agent.

```python
sold_properties = df[df['status'] == 'Sold']

df['total_commision'] = sold_properties['sale_price'] * sold_properties['agent_commission']


agent_commision = df.groupby('agent_name')['total_commision'].sum()


agent_commision.plot(kind='bar')
plt.title('Total Commision Earned by Agent')
plt.xlabel('Agent Name')
plt.ylabel('Total Commision')
plt.xticks(rotation=0)
plt.show()
```

Problem 21: NumPy Statistics
Using NumPy, calculate the mean, median, and standard deviation of sale_price (for sold properties).

Problem 23: Price Category
Create a new column price_category based on list_price:

"Budget" if < 5,000,000
"Mid-Range" if 5,000,000 - 15,000,000
"Luxury" if > 15,000,000

```python
df['price_category'] = df['sale_price'].apply(
    lambda x: 'Budget' if x < 5_000_000
    else 'Mid-Range' if x <= 14_999_999
    else 'Luxury'
)


df[['property_type','price_category','sale_price']].head(5)




```
