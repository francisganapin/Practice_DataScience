```python
import pandas as pd

df = pd.read_csv('real_estate_data.csv')



df['commission_earned'] = df['sale_price'] * df['commission_rate']


agent_performance = df.groupby(['agent_id','agent_name']).agg({
    'commission_earned':['sum','mean','count']
}).round(2)


agent_performance.columns = ['Total Commision','Avg Commision','Properties Sold']
agent_performance = agent_performance.sort_values('Total Commision',ascending=False)

agent_performance.head(10)


```

```python
df['discount_pct'] = ((df['listing_price'] - df['sale_price']) / df['listing_price']) * 100


property_analysis = df.groupby('property_type').agg({
    'discount_pct':'mean',
    'sale_price':'mean',
    'property_id':'count'
}).round(2)

property_analysis.columns = ['Avg Discount','Avg Sale Price','Properties Sold']
property_analysis = property_analysis.sort_values('Avg Discount',ascending=False)
property_analysis.head(10)

```

```python
property_analysis.columns = ['Avg Discount%','Avg Sale Price','Count']
property_analysis = property_analysis.sort_values('Avg Discount%',ascending=False)
property_analysis.head(10)
```

```python
df['discount_amount'] = df['listing_price'] * df['sale_price']



df['market_time_category'] = pd.cut(
    df['days_on_market'],
    bins=[0,30,45,60,100],
    labels=['Quick (0-30)', 'Normal (31-45)', 
                                            'Slow (46-60)', 'Very Slow (60+)'])

time_analysis = df.groupby('market_time_category').agg({
    'discount_pct':'mean',
    'discount_amount':'mean',
    'property_id':'count'
}).round(2)


print(time_analysis)

correlation = df['days_on_market'].corr(df['discount_pct'])
print(f'Correlation between days on market and discount percentage: {correlation}')


```

```python
neighborhood_stats = df.groupby('neighborhood').agg({
    'sale_price':['sum','mean','count'],
    'commission_rate':'mean'
}).round(2)

neighborhood_stats

```

```python
df['years_since_renovation'] = 2024 - df['last_renovation_year']

df['renovation_category'] = pd.cut(df['years_since_renovation'],
    bins=[-1,3,6,10,50],
    labels=['Recent(0-3 yrs)', 'Moderate(4-6)',
     'Older (7-10)','Very Old(10+yrs)'])


df['renovation_category'].head(1)
```

```python
renovation_analysis = df.groupby('renovation_category').agg({
    'sale_price':'mean',
    'discount_pct':'mean',
    'days_on_market':'mean',
    'property_id':'count'
}).round(2)
```

```python
renovation_analysis
```

```python
renovation_analysis.columns = ['Avg Sale Price', 'Avg Discount %', 'Avg Days on Market', 'Count']
renovation_analysis
```

```python
recent_reno = df[df['years_since_renovation'] < 3]['sale_price'].mean()
old_reno = df[df['years_since_renovation'] > 3]['sale_price'].mean()

print(f'Price Premium for Recent Renovation: ${recent_reno - old_reno:.2f}')
```

```python
agent_efficiency = df.groupby(['agent_id','agent_name']).agg({
    'days_on_market':['mean','median'],
    'sale_price':'mean',
    'property_id':'count',
    'commission_earned':'sum',
    
}).round(2)

agent_efficiency.columns = ['mean_days_on_market','median_days_on_market','mean_sale_price','property_count','total_commission_earned']



agent_efficiency['efficiency_score'] = (
    agent_efficiency['property_count'] / agent_efficiency['mean_days_on_market'] * 100
).round(2)



agent_efficiency = agent_efficiency.sort_values('efficiency_score',ascending=False)
print('Agent Efficiency Score')
print(agent_efficiency.head())
```

```python
df['price_per_sqft'] = (df['sale_price'] / df['square_feet']).round(2)

df['price_per_sqft'].head()





```

```python
df['size_category'] = pd.cut(df['square_feet'],
                    bins=[0,1200,1800,2500,5000],
                    labels=['small','medium','large','extra_large'],
                    include_lowest=True
)

size_analysis = df.groupby('size_category').agg({
    'price_per_sqft':'mean',
    'sale_price':'mean',
    'square_feet':'mean',
    'days_on_market':'mean',
    'property_id':'count'
}).round(2)

size_analysis.columns = ['Avg Price/SqFt','Avg Sale Price','Avg Square Feet',
'Avg Square Feet','Avg Days on Market']

print('Pricing Analysis by Size Category')
print(size_analysis)



detailed_analysis = (
    df.groupby(['property_type','size_category'])[
        ['price_per_sqft','sale_price','days_on_market']
    ]
    .mean()
    .round(2)
)



print('nPrice per SqFt by Type and Size:')
print(detailed_analysis.unstack())



```
