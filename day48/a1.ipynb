```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employees.csv')
```

```python
#How many rows and columns are in the dataset?

column = df.shape[0]   # actually rows
rows = df.shape[1]    # actually columns
print("Number of rows: ", rows)
print("Number of columns: ", column)
```

```python
#data type for each column
df.dtypes
```

```python
#Are there any missing values in the dataset?

df.isnull().sum()
```

```python
#What is the average salary of all employees?

average_salary_per_department = df.groupby('Department')['Salary'].mean().round(2)
print(average_salary_per_department)
```

```python
#Filter the data to show only employees in the 'Engineering' departmen

employee_engineering = df[df['Department'] == 'Engineering']
employee_engineering
```

```python
employee_earn_more_than_100 = df[df['Salary'] > 100000]
print(employee_earn_more_than_100[['FirstName','LastName','Salary']])
```

```python
#Find all employees who joined after January 1, 2020.

employee_who_joined_after_jan_2020 = df[df['JoiningDate'] > '2020-01-01']
employee_who_joined_after_jan_2020
```

```python
#Who is the oldest employee in the company

employee_olders = df['Age'].nlargest(1)
print(df.iloc[employee_olders.index])
```

```python
#List all employees who work in 'San Francisco'

employee_sf = df[df['City'] == 'San Francisco']
print(employee_sf[['FirstName','LastName','City']])
```

```python
salary_per_department = df.groupby('Department').agg(
    average_salary_of_department=('Salary','mean')
).sort_values(
    by='average_salary_of_department',
    ascending=False
).reset_index()


salary_per_department
```

```python
#How many employees are there in each city

nubmer_employee_per_city = df.groupby('City').agg(
    count = ('EmployeeID','count')
).sort_values(
    by='count',
    ascending=False
).reset_index()

nubmer_employee_per_city
```

```python
#what is the maximum performance score in the 'Sales' department
```

```python
value = df[df['Department'] == 'Sales']['PerformanceScore'].max()

print(df[(df['Department'] == 'Sales') & (df['PerformanceScore'] == value)])
```

```python
#What is the average age of employees in each department?

average_age_per_department = df.groupby('Department').agg(
    avg_age = ('Age','mean'),
    avg_salary = ('Salary','mean')
).sort_values(
    by='avg_age',
    ascending=False
).reset_index().round(2)

average_age_per_department
```

```python

plt.figure(figsize=(16,10))  # set figure size BEFORE plotting
average_age_per_department.plot(kind='bar', x='Department', y='avg_age')
plt.xticks(rotation=0)
plt.show()
```

```python
#Create a new column called 'Seniority' that calculates the number of years each employee has been with the company (assuming current year is 2024)

df['JoiningDate'] = pd.to_datetime(df['JoiningDate'])
df['years_of_experiance'] =   2024 - df['JoiningDate'].dt.year



df['seniority'] = np.where(df['years_of_experiance'] >= 5, 'Senior', 'Junior')
df


```
