```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Employee': ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'coffee_cups': [2, 4, 3, 5, 1],
    'hours_worked': [8, 9, 7, 10, 6],
    'sales': [500, 600, 450, 700, 400],
    'breaks': [2, 1, 3, 1, 4]
})
```

```python
corr = df[['coffee_cups','hours_worked','sales','breaks']].corr()
print(corr)
```

```python
corr_cuffee_worked = df[['coffee_cups','hours_worked']].corr()
corr_cuffee_worked
```

```python
from sklearn.linear_model import LinearRegression

X = df[['coffee_cups']]
y = df['sales']

model = LinearRegression()
model.fit(X,y)

print("Sales increase per coffee cup:", model.coef_[0])
print("Base sales:", model.intercept_)
```
