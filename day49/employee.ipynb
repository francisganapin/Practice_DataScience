```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('employee_data.csv')
df.head()
```

```python
average_salary = df['salary'].mean()
print(f"Average salary: {average_salary} of all employee")
```

```python
average_salary_per_employee =  df.groupby('Employee_ID')['salary'].mean()
average_salary_per_employee.plt(kind='bar')
```

```python
plt.plot(average_salary_per_employee)
plt.xlabel("Employee ID")
plt.ylabel("Salary")
plt.title("Average Salary per Employee")
plt.show()
```
