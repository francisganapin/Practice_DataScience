# Employee SQL Tutorial for Beginners

Welcome! This notebook will teach you SQL using an employee database.

## Database Overview

We have 4 tables:
- **departments** - Company departments
- **employees** - Employee information
- **projects** - Company projects
- **employee_projects** - Who works on which project

---

## Setup - Run this first!

```python
import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect('employee.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("Connected to employee database!")
```

## Let's Look at Our Tables First

```python
# View the departments table
run_sql("SELECT * FROM departments")
```

```python
# View the employees table
run_sql("SELECT * FROM employees")
```

```python
# View the projects table
run_sql("SELECT * FROM projects")
```

```python
# View the employee_projects table
run_sql("SELECT * FROM employee_projects")
```

## Problem 7: COUNT Employees per Department
**Question:** How many employees are in each department?

**Hint:** Use `COUNT(*)` with `GROUP BY`

```python
#Write your answer here:
run_sql("""
    SELECT  d.department_id, department_name, COUNT(*) as count_per_department
    from employees s
    JOIN departments d
        ON d.department_id = s.department_id
    GROUP BY department_name

""")
```

## Problem 15: Multiple JOINs - Employee Project Details
**Question:** Show employee names, project names, and their roles on each project.

**Hint:** Join `employees`, `employee_projects`, and `projects` tables

```python
#Write your answer here:
run_sql("""
    SELECT e.first_name,e.last_name,ep.role,p.project_name
    from employees e
    JOIN employee_projects ep
        on e.employee_id = ep.employee_id
    JOIN projects p
        on ep.project_id = p.project_id
    
""")
```

---
## Problem 16: HAVING - Filter Groups
**Question:** Find departments that have more than 2 employees.

**Hint:** Use `GROUP BY` with `HAVING COUNT(*) > 2`

```python
#Write your answer here:
run_sql("""
    SELECT department_id,count(*) as employee_count
    from employees
    group by department_id
    having count(*) > 2
""")
```
