## Setup - Run this first!

```python
import sqlite3
import pandas as pd

# Connect to database
conn = sqlite3.connect('employee.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("âœ… Connected to employee database!")
```

## Problem 15: Multiple JOINs - Employee Project Details
**Question:** Show employee names, project names, and their roles on each project.

ðŸ’¡ **Hint:** Join `employees`, `employee_projects`, and `projects` tables

```python
# âœï¸ Write your answer here:
run_sql("""

""")
```

### âœ… Answer 15

```python
# Answer:
run_sql("""
SELECT e.first_name, e.last_name, p.project_name, ep.role, ep.hours_worked
FROM employees e
JOIN employee_projects ep ON e.employee_id = ep.employee_id
JOIN projects p ON ep.project_id = p.project_id
ORDER BY p.project_name, e.last_name
""")

# Expected: 15 rows showing who works on which project with their role
```

## Problem 16: HAVING - Filter Groups
**Question:** Find departments that have more than 2 employees.

ðŸ’¡ **Hint:** Use `GROUP BY` with `HAVING COUNT(*) > 2`

```python
# âœï¸ Write your answer here:
run_sql("""

""")
```

### âœ… Answer 16

```python
# Answer:
run_sql("""
SELECT department_id, COUNT(*) as employee_count
FROM employees
GROUP BY department_id
HAVING COUNT(*) > 2
""")

# Expected: Department 1 (5 employees), Department 2 (3 employees), Department 5 (3 employees)
```
