```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dirty_employees.csv', 
                 na_values=['NULL', 'N/A', ''],  # Treat as NaN
                 parse_dates=['hire_date'])



```

```python
#find missing number

df.isnull().sum()
```

```python
#total number of missingn values

df.isnull().sum().sum()
```

```python
#percentage of missing
percent_missing = df.isnull().sum() * 100 / len(df)
print(percent_missing)
```

```python
#replace placeholder text with Nan


df.replace(['NULL','N/A','n/a','',' '],np.nan,inplace=True)
df




```

```python
df['salary'].fillna(df['salary'].mean(),inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
#view duplicate rows
df[df.duplicated(keep=False)]

```

```python
df
```

```python
df['first_name'] = df['first_name'].str.lower()



```

```python
df
```

```python
df['last_name'] = df['last_name'].str.lower()
```

```python
df[['last_name', 'first_name']] = df[['last_name','first_name']].apply(lambda x: x.str.capitalize(), axis=1)
```

```python
df
```

```python
df['salary'] = pd.to_numeric(df['salary'],errors='coerce')
df.head(5)
```

```python
from scipy import stats
```

```python
z_scores = stats.zscore(df['salary'])
df = df[(np.abs(z_scores) < 2)]
df
```
