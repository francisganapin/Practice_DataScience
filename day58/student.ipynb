```python
#fixing the score not align
df['Math_Score'] = df['Math_Score'].clip(upper=100)
df['Science_Score'] = df['Science_Score'].clip(upper=100)
df['English_Score'] = df['English_Score'].clip(upper=100)


```

```python
print("---- Suspicious Math Score ----")
suspicious_math_score = df[df['Math_Score'] > 100]
print(suspicious_math_score)
```

```python
df.isnull().sum()
```

```python
#fill the missing value by mean
df['Study_Hours'] = df['Study_Hours'].fillna(df['Study_Hours'].mean())
# check again if thier is no missing value
df.isnull().sum()
```

```python
#Feature Engineering (Creating new Column)
#people can be engineer
df['Overall_Score'] = (df['Math_Score'] + df['Science_Score'] + df['English_Score']) / 3
df['Overall_Score'] = df['Overall_Score'].round(2)

```

```python
top_students = df.sort_values(by='Overall_Score',ascending=False)
print(top_students[['Student_ID','Overall_Score','Has_Part_Time_Job']].head(5))
```
