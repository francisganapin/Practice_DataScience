```python
daily_views = df['views']
cumulative_views = daily_views.cumsum()

plt.plot(cumulative_views)
plt.title('Total Views Growth over Time')
plt.show()
```

```python
#filtering data
views_10k = df[df['views'] > 10000]

print("*" * 500)
display(views_10k)
print('Display all 10k views')
print("*" * 500)

tiktok_or_youtube = df[(df['platform_num'] == 'Tiktok') | (df['platform_num'] == 'Youtube')]

print("*" * 500)
display(tiktok_or_youtube)
print('Show tiktok or Youtube')
print("*" * 500)
```

```python
platform_stats = df.groupby('platform_num').agg({
    'views':['mean','sum','count'],
    'likes':['mean','sum'],
    'engagement_rate':'mean'
}).round(2)
display(platform_stats)
```
