```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('real_estate_transactions.xlsx',sheet_name='Transactions')
```

```python
df['price_ratio'] = round(df['sold_price'] / df['listed_price'],4)
```

```python
df.head(3)
```

```python
df['flag_price_flip'] = df['price_ratio'] > 1.3
```

```python
df.head(3)
```

```python
df['value_ratio'] = round(df['sold_price'] / df['appraised_value'],4)
df['flag_undervalue'] = df['value_ratio'] < 0.60
```

```python
df.head(3)
```

```python
df['zscore_sold'] = df.groupby("property_type")["sold_price"].transform(
    lambda x: (x - x.mean()) / x.std()
).round(4)
```

```python
df['flag_zscore_outlier'] = df['zscore_sold'].abs() > 2
```

```python
df.head(2)
```

```python
df['sold_speed'] = pd.cut(
    df['days_on_market'],
    bins=[1,70,110,float('inf')],
    labels=['fast','average','slow']
)
```

```python
df.head(3)
```

```python
fast = df[df['sold_speed'] == 'average']
fast
```
