# Do total debits equal total credits? Prove it.

```python
debit = df['debit'].sum()
credit = df['credit'].sum()


if debit == credit:
    print(f'Total debit {debit} and credit are both equal {credit}')
else:
    print(f'they are not equal debit was {debit} and credit {credit}')

```

# Build a Trial Balance showing each account's total debit, total credit, and balance.

```python
trial_balance = df.groupby('account')[['debit','credit']].sum().reset_index()
trial_balance['balance'] = trial_balance['debit'] - trial_balance['credit']

print(trial_balance)
```
