# **Net Profit**: Kumita ka ba o nalugi? (Kita - Gastos)


```python
kumitaba = kabuuang_kita - kabuuang_gatos

if kumitaba >= 0:
    print(f'hindi tayo nalugi {kumitaba}')
else:
    print(f'Nalugi tayo  {kumitaba}')
```

```python
if kabuuang_kita > kabuuang_gatos:
    print(f'Kumita tayo {kumitaba}')
elif kabuuang_kita == kabuuang_gatos:
    print(f'Print break even tayo {kumitaba}')
else:
    print(f'Nalugi tayo {kumitaba}')
```

**Hawaan ng Pera (Cash on Hand)**: Magkano lang ang _Actual Cash_ na hawak mo ngayon?

```python
df3 = pd.read_csv('talaarawan_v2.csv')
```

```python
df3.head(3)
```

```python
kita = df3[ (df3['Uri'] == 'KITA') & (df3['Payment_Type'] == 'CASH') & (df3['Status'] == 'PAID') ]['Halaga'].sum()
gastos =  df3[ (df3['Uri'] == 'GASTOS') & (df3['Payment_Type'] == 'CASH') & (df3['Status'] == 'PAID') ]['Halaga'].sum()

total_cash_in_hand = kita - gastos

print(f'Ang total cash natin in hand ay {total_cash_in_hand}')


```

# **GCash Balance**: Magkano ang laman ng GCash wallet mo base sa records?

```python
kita_g = df3[ (df3['Uri'] == 'KITA') & (df3['Payment_Type'] == 'GCASH') & (df3['Status'] == 'PAID') ]['Halaga'].sum()
gastos_g =  df3[ (df3['Uri'] == 'GASTOS') & (df3['Payment_Type'] == 'GCASH') & (df3['Status'] == 'PAID') ]['Halaga'].sum()

laman_gcash = kita_g - gastos_g

print(f'Ang total cash natin in GCASH ay {laman_gcash}')
```

**Maniningil (Accounts Receivable)**: Magkano pa ang **Total Utang** na hindi pa nababayaran ng mga customer?

```python
total_utang_ng_customer = df3[(df3['Uri'] == 'KITA') & (df3['Status'] == 'UTANG')]['Halaga'].sum()

print(f'{total_utang_ng_customer} ito ay kulang pa ng mga ating customer')



```
